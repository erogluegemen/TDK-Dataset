```python
import ast
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('tdk_word_data_all.csv')
df.head(3)
```

```python
df = df[['madde', 'anlamlarListe']]
df.head(3)
```

```python
df['anlamlarListe'].fillna('[]', inplace=True)
df['anlamlarListe'] = df['anlamlarListe'].apply(ast.literal_eval)
anlam_list = df['anlamlarListe'].apply(lambda x: x[0]['anlam'] if x else 'No Data')
df['anlam'] = anlam_list
df.drop('anlamlarListe', axis=1, inplace=True)

df.head(3)
```

```python
df.head(10)
```

```python
df.to_csv('tdk_word_meaning_data.csv', index=False)
```
